```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score,f1_score,roc_curve, precision_recall_curve, PrecisionRecallDisplay

from sklearn import metrics
from sklearn.model_selection import train_test_split

def aqi_pm25(a):
    pm_concentration = a
    range_of_aqi = [0.0,50.0,100.0,300,500]
    concentration_range = [0,12,35.5,250.5,500.5]
    if pm_concentration > concentration_range[0] and pm_concentration <= concentration_range[1]:
        return concentration_range[0],concentration_range[1],range_of_aqi[0],range_of_aqi[1]
    elif pm_concentration > concentration_range[1] and pm_concentration <= concentration_range[2]:
        return concentration_range[1],concentration_range[2],range_of_aqi[1],range_of_aqi[2]
    elif pm_concentration > concentration_range[2] and pm_concentration <= concentration_range[3]:
        return concentration_range[2],concentration_range[3],range_of_aqi[2],range_of_aqi[3]
    elif pm_concentration > concentration_range[3] and pm_concentration <= concentration_range[4]:
        return concentration_range[3],concentration_range[4],range_of_aqi[3],range_of_aqi[4]
    else:
        return concentration_range[3],concentration_range[4],range_of_aqi[3],range_of_aqi[4]

def calculate_aqi_pm25(a):
    current_aqi_value = 0
    current_concentration_value = a
    low_concentration,high_concentration,low_aqi_value,high_aqi_value = aqi_pm25(current_concentration_value)
    current_aqi_value = (( high_aqi_value - low_aqi_value ) * ( current_concentration_value - low_concentration ) / (high_concentration - low_concentration)) + (low_aqi_value)
    return current_aqi_value

def aqi_pm10(a):
    pm_concentration = a
    range_of_aqi = [0.0,50.0,100.0,300.0,500]
    concentration_range = [0,55,155,425,605]
    if pm_concentration > concentration_range[0] and pm_concentration <= concentration_range[1]:
        return concentration_range[0],concentration_range[1],range_of_aqi[0],range_of_aqi[1]
    elif pm_concentration > concentration_range[1] and pm_concentration <= concentration_range[2]:
        return concentration_range[1],concentration_range[2],range_of_aqi[1],range_of_aqi[2]
    elif pm_concentration > concentration_range[2] and pm_concentration <= concentration_range[3]:
        return concentration_range[2],concentration_range[3],range_of_aqi[2],range_of_aqi[3]
    elif pm_concentration > concentration_range[3] and pm_concentration <= concentration_range[4]:
        return concentration_range[3],concentration_range[4],range_of_aqi[3],range_of_aqi[4]
    else:
        return concentration_range[3],concentration_range[4],range_of_aqi[3],range_of_aqi[4]

def calculate_aqi_pm10(a):
    current_aqi_value = 0
    current_concentration_value = a
    low_concentration,high_concentration,low_aqi_value,high_aqi_value = aqi_pm10(current_concentration_value)
    current_aqi_value = (( high_aqi_value - low_aqi_value ) * ( current_concentration_value - low_concentration ) / (high_concentration - low_concentration)) + (low_aqi_value)
    return current_aqi_value

def aqi_no2(a):
    pm_concentration = a
    range_of_aqi = [0.0,50.0,100.0,300.0,500.0]
    concentration_range = [0,54,101,1250,2049]
    if pm_concentration > concentration_range[0] and pm_concentration <= concentration_range[1]:
        return concentration_range[0],concentration_range[1],range_of_aqi[0],range_of_aqi[1]
    elif pm_concentration > concentration_range[1] and pm_concentration <= concentration_range[2]:
        return concentration_range[1],concentration_range[2],range_of_aqi[1],range_of_aqi[2]
    elif pm_concentration > concentration_range[2] and pm_concentration <= concentration_range[3]:
        return concentration_range[2],concentration_range[3],range_of_aqi[2],range_of_aqi[3]
    elif pm_concentration > concentration_range[3] and pm_concentration <= concentration_range[4]:
        return concentration_range[3],concentration_range[4],range_of_aqi[3],range_of_aqi[4]
    else:
        return concentration_range[3],concentration_range[4],range_of_aqi[3],range_of_aqi[4]

def calculate_aqi_no2(a):
    current_aqi_value = 0
    current_concentration_value = a
    low_concentration,high_concentration,low_aqi_value,high_aqi_value = aqi_no2(current_concentration_value)
    current_aqi_value = (( high_aqi_value - low_aqi_value ) * ( current_concentration_value - low_concentration ) / (high_concentration - low_concentration)) + (low_aqi_value)
    return current_aqi_value

def aqi_co(a):
    pm_concentration = a
    range_of_aqi = [0.0,50.0,100.0,300.0,500.0]
    concentration_range = [0,4.5,9.5,30.5,50.5]
    if pm_concentration > concentration_range[0] and pm_concentration <= concentration_range[1]:
        return concentration_range[0],concentration_range[1],range_of_aqi[0],range_of_aqi[1]
    elif pm_concentration > concentration_range[1] and pm_concentration <= concentration_range[2]:
        return concentration_range[1],concentration_range[2],range_of_aqi[1],range_of_aqi[2]
    elif pm_concentration > concentration_range[2] and pm_concentration <= concentration_range[3]:
        return concentration_range[2],concentration_range[3],range_of_aqi[2],range_of_aqi[3]
    elif pm_concentration > concentration_range[3] and pm_concentration <= concentration_range[4]:
        return concentration_range[3],concentration_range[4],range_of_aqi[3],range_of_aqi[4]
    else:
        return concentration_range[3],concentration_range[4],range_of_aqi[3],range_of_aqi[4]

def calculate_aqi_co(a):
    current_aqi_value = 0
    current_concentration_value = a
    low_concentration,high_concentration,low_aqi_value,high_aqi_value = aqi_co(current_concentration_value)
    current_aqi_value = (( high_aqi_value - low_aqi_value ) * ( current_concentration_value - low_concentration ) / (high_concentration - low_concentration)) + (low_aqi_value)
    return current_aqi_value

def aqi_so2(a):
    pm_concentration = a
    range_of_aqi = [0.0,50.0,100.0,150.0,200.0]
    concentration_range = [0,36,76,186,304]
    if pm_concentration > concentration_range[0] and pm_concentration <= concentration_range[1]:
        return concentration_range[0],concentration_range[1],range_of_aqi[0],range_of_aqi[1]
    elif pm_concentration > concentration_range[1] and pm_concentration <= concentration_range[2]:
        return concentration_range[1],concentration_range[2],range_of_aqi[1],range_of_aqi[2]
    elif pm_concentration > concentration_range[2] and pm_concentration <= concentration_range[3]:
        return concentration_range[2],concentration_range[3],range_of_aqi[2],range_of_aqi[3]
    elif pm_concentration > concentration_range[3] and pm_concentration <= concentration_range[4]:
        return concentration_range[3],concentration_range[4],range_of_aqi[3],range_of_aqi[4]
    else:
        return concentration_range[3],concentration_range[4],range_of_aqi[3],range_of_aqi[4]

def aqi_o3(a):
    pm_concentration = a
    range_of_aqi = [0.0,100.0,300.0,500.0]
    concentration_range = [0,125,405,605]
    if pm_concentration > concentration_range[0] and pm_concentration <= concentration_range[1]:
        return concentration_range[0],concentration_range[1],range_of_aqi[0],range_of_aqi[1]
    elif pm_concentration > concentration_range[1] and pm_concentration <= concentration_range[2]:
        return concentration_range[1],concentration_range[2],range_of_aqi[1],range_of_aqi[2]
    elif pm_concentration > concentration_range[2] and pm_concentration <= concentration_range[3]:
        return concentration_range[2],concentration_range[3],range_of_aqi[2],range_of_aqi[3]
    else:
        return concentration_range[2],concentration_range[3],range_of_aqi[2],range_of_aqi[3]

def calculate_aqi_o3(a):
    current_aqi_value = 0
    current_concentration_value = a
    low_concentration,high_concentration,low_aqi_value,high_aqi_value = aqi_o3(current_concentration_value)
    current_aqi_value = (( high_aqi_value - low_aqi_value ) * ( current_concentration_value - low_concentration ) / (high_concentration - low_concentration)) + (low_aqi_value)
    return current_aqi_value

def calculate_aqi_so2(a):
    current_aqi_value = 0
    current_concentration_value = a
    low_concentration,high_concentration,low_aqi_value,high_aqi_value = aqi_so2(current_concentration_value)
    current_aqi_value = (( high_aqi_value - low_aqi_value ) * ( current_concentration_value - low_concentration ) / (high_concentration - low_concentration)) + (low_aqi_value)
    return current_aqi_value

def calculate_overall_aqi(pm25,pm10,so2,no2,co,o3):
    range_of_aqi = [0.0,50.0,100.0,300.0,500.0]
    max_aqi = max(pm25,pm10,so2,no2,co,o3)
    if max_aqi>=range_of_aqi[0] and max_aqi<range_of_aqi[1]:
        return 1
    elif max_aqi>=range_of_aqi[1] and max_aqi<range_of_aqi[2]:
        return 2
    elif max_aqi>=range_of_aqi[2] and max_aqi<range_of_aqi[3]:
        return 3
    else:
        return 4

def convert_measurements(df):
    df['SO2'] = df['SO2']/2.62
    df['NO2'] = df['NO2']/1.88
    df['O3'] = df['O3']/2.00
    df['CO'] = df['CO']/(1.145*1000)
df = pd.read_csv('/content/PRSA_Data_Aotizhongxin_20130301-20170228.csv')
df.dropna()
convert_measurements(df)
df['PM25_AQI'] = 0
print(df.head(1))
for i in range(len(df)):
    a = df['PM2.5'].iloc[i]
    df['PM25_AQI'].iloc[i] = calculate_aqi_pm25(a)
df['PM10_AQI'] = 0
for i in range(len(df)):
    a = df['PM10'].iloc[i]
    df['PM10_AQI'].iloc[i] = calculate_aqi_pm10(a)
df['SO2_AQI'] = 0
for i in range(len(df)):
    a = df['SO2'].iloc[i]
    df['SO2_AQI'].iloc[i] = calculate_aqi_so2(a)
df['NO2_AQI'] = 0
for i in range(len(df)): 
    a = df['NO2'].iloc[i]
    df['NO2_AQI'].iloc[i] = calculate_aqi_no2(a)
df['CO_AQI'] = 0
for i in range(len(df)):
    a = df['CO'].iloc[i]
    df['CO_AQI'].iloc[i] = calculate_aqi_co(a)
df['O3_AQI'] = 0
for i in range(len(df)):
    a = df['O3'].iloc[i]
    df['O3_AQI'].iloc[i] = calculate_aqi_o3(a)
df['Overall_AQI'] = 0
```

```python
df = df.drop(['No'],axis=1)
df = df.drop(['year'],axis=1)
df = df.drop(['month'],axis=1)
df = df.drop(['day'],axis=1)
df = df.drop(['hour'],axis=1)
df = df.drop(['TEMP'],axis=1)
df = df.drop(['PRES'],axis=1)
df = df.drop(['DEWP'],axis=1)
df = df.drop(['RAIN'],axis=1)
df = df.drop(['wd'],axis=1)
df = df.drop(['WSPM'],axis=1)
df = df.drop(['station'],axis=1)

for i in range(len(df)):
    df['Overall_AQI'].iloc[i] = calculate_overall_aqi(df['PM25_AQI'].iloc[i],df['PM10_AQI'].iloc[i],df['SO2_AQI'].iloc[i],df['NO2_AQI'].iloc[i],df['CO_AQI'].iloc[i],df['O3_AQI'].iloc[i])
```

Sklearn Linear Regression Implementation

```python
df = df.dropna()
X = df.drop(['Overall_AQI'], axis=1)
y = df['Overall_AQI'] 
X_test,X_train,  y_test, y_train = train_test_split(X, y, test_size=0.2, random_state=0)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print('Root Mean Squared Error: ', rmse)
print('R2 Score: ', r2)
```

```python
y_pred = (np.rint(y_pred)).astype(int)
y_pred[y_pred>4] = 4
print(y_pred)
print(y_test)
```

```python
def confusion_matrix(y, y_pred, classes,fig,class_type):
    if class_type == "Binary":
      confusion_matrix = np.zeros((2, 2))
      rows = ["Actual Positive", "Actual Negative"]
      cols = ("Classifier Positive", "Classifier Negative")
    else:
      confusion_matrix = np.zeros((len(classes),len(classes)))
      print(confusion_matrix.shape)
      rows = []
      cols = []
      for cl in classes:
          rows.append("Actual " + str(cl))
          cols.append("Predicted " + str(cl))
    for i, j in zip(y, y_pred):
        try:
          confusion_matrix[i-1][j-1] += 1
        except:
           print(i,j)
    fig.subplots_adjust(left=0.3,top=0.8, wspace=2)
    ax = plt.subplot2grid((1,1), (0,0), colspan=2, rowspan=2)
    table = ax.table(cellText=confusion_matrix.tolist(),
          rowLabels=rows,
          colLabels=cols, loc="upper center")
    table.set_fontsize(14)
    table.scale(5, 6)
    ax.axis("off")
fig = plt.figure(1)
classes = ["Healthy","Moderate","Unhealthy","Hazardous"]
confusion_matrix(y_test,y_pred,classes,fig,"multi")
print("F1 Score:"+f1_score(y_test,y_pred,average="weighted").astype(str))
for l in [1,2,3,4]:
  est = ""
  if l ==1:
     est = "Healthy"
  elif l ==2:
     est = "Moderate"
  elif l == 3:
     est = "Unhealthy"
  else:
     est = "Hazardous"
  tpr,fpr,thersholds = metrics.roc_curve(y_test,y_pred,pos_label = l)
  roc_auc = metrics.auc(fpr, tpr)
  display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,estimator_name=est)
  display.plot()
  plt.show()

for l in [1,2,3,4]:

  est = ""
  if l ==1:
     est = "Healthy"
  elif l ==2:
     est = "Moderate"
  elif l == 3:
     est = "Unhealthy"
  else:
     est = "Hazardous"

  precision, recall, _ = precision_recall_curve(y_test, y_pred,pos_label=l)
  disp = PrecisionRecallDisplay(precision=precision, recall=recall,estimator_name=est)
  disp.plot()
  plt.show()
```
